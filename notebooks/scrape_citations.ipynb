```python
from bs4 import BeautifulSoup
import requests, json
import pandas as pd
```

```python
experts = pd.read_csv('twitter_accounts_2.csv')
experts['name'] = experts['name'].apply(lambda x: x.split('@')[0].split(',')[0].strip())
experts = experts[['name', 'twitter', 'institution']].copy()
```

```python
updated_dict = json.load(open('cleaned_account_info.json')) 
experts['desc'] = experts['twitter'].apply(lambda x: 
                                           updated_dict[x]['desc'] if x in updated_dict else '')
#experts.to_csv('twitter_accounts_desc_citation.csv', index=False)
```

```python
names = list(experts['name'])
cite, results = [], []
for target in names: 
    target = [target.split(' ')[0], target.split(' ')[-1]]
    search_name = '+'.join(target).lower()
    url = f'https://scholar.google.com/citations?view_op=search_authors&mauthors={search_name}'
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser') 

    name = [i['alt'] for i in soup.findAll("img")]
    desc = [i.text for i in soup.findAll("div", {'class':['gs_ai_aff']})]
    citation = [i.text for i in soup.findAll("div", {'class':['gs_ai_cby']})]
    
    result = list(zip(name, desc, citation))
    results.append(result)
    cite.append(len(result))
    #print(target, result)
```

```python
experts['cite'] = cite
experts['citation'] = results
```

```python
experts.to_csv('twitter_accounts_desc_citation.csv', index=False)
```
